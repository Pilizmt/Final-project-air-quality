```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    print('Dickey-Fuller test results:')
    dftest = adfuller(timeseries, autolag = 'AIC')
    dfoutput = pd.Series(dftest[0:4], index = ['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return dfoutput

test_stationarity(ts_df)
```

**Es estacionaria**

```python
from pmdarima import auto_arima

# Supongamos que tienes una serie de tiempo en 'ts_df'

# Realiza la b√∫squeda autom√°tica de hiperpar√°metros
model = auto_arima(ts_df, seasonal=True, trace=True,
                   start_p=0, max_p=3,
                   start_q=0, max_q=3,
                   start_P=0, max_P=2,
                   start_Q=0, max_Q=2,
                   d=1, D=1, m=12,  # Aqu√≠, 'm' representa la periodicidad estacional (por ejemplo, 12 para datos mensuales)
                   stepwise=True)

# Obtiene los hiperpar√°metros recomendados
p, d, q = model.order
P, D, Q, s = model.seasonal_order

print(f"Mejores hiperpar√°metros encontrados:")
print(f"Orden ARIMA: ({p}, {d}, {q})")
print(f"Orden Estacional: ({P}, {D}, {Q}, {s})")
```
