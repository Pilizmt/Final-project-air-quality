# **Danger Graphs by Date**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_date = pd.read_csv('../data/processed/clean_M_date.csv')
```

```python
df_date.head()
```

```python
# Convert column 'FECHA' to 'datetime'
df_date['FECHA'] = pd.to_datetime(df_date['FECHA'])
```

```python
# Create a line chart with different colours for each hazard level (Full Year)

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_date, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("Full Year")
plt.ylabel("Pollutant Level")
plt.title("Evolution of Pollutants by Danger Level Over Time")

plt.xticks([])
plt.legend(title='Danger Level')
plt.show()
```

```python
# Create a line graph for the average pollutants per hazard level (First half of the year)

start_date_6 = pd.to_datetime('2022-01-01')
end_date_6 = pd.to_datetime('2022-06-30')

date_range_data_6 = df_date[(df_date['FECHA'] >= start_date_6) & (df_date['FECHA'] <= end_date_6)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data_6, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("First half of the year")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

```python
# Create a line graph for the average pollutants per hazard level (March and April)

start_date = pd.to_datetime('2022-03-01')
end_date = pd.to_datetime('2022-04-30')

date_range_data = df_date[(df_date['FECHA'] >= start_date) & (df_date['FECHA'] <= end_date)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("March and April dates")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

```python
# Create a line graph for the average pollutants per hazard level (8/3/2022 to 22/3/2022)

start_date = pd.to_datetime('2022-03-08')
end_date = pd.to_datetime('2022-03-22')

date_range_data = df_date[(df_date['FECHA'] >= start_date) & (df_date['FECHA'] <= end_date)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("Date from 8/3/2022 to 22/3/2022")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

La contaminación en Madrid el 15 de marzo de 2022 fue bastante alta.

Algunas de las posibles causas de esta situación son:

- El aumento del tráfico rodado debido al fin de las restricciones por la pandemia de COVID-19.
- La falta de lluvias y viento que favorecen la dispersión de los contaminantes.
- La emisión de gases de efecto invernadero por la quema de combustibles fósiles, como el carbón, el petróleo y el gas natural.
- Se registró un pico de partículas en suspensión (PM10) que superó el valor límite diario de 50 µg/m32.
- Hubo una situación de anticiclón que impidió la dispersión de los contaminantes.
- Se registró una entrada de polvo sahariano que contribuyó a elevar los niveles de partículas.

```python
# Create a line graph for the average pollutants per hazard level (July to September)

start_date = pd.to_datetime('2022-07-01')
end_date = pd.to_datetime('2022-09-30')

date_range_data = df_date[(df_date['FECHA'] >= start_date) & (df_date['FECHA'] <= end_date)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("Dates from June to September")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

La contaminación en los meses de entre julio y septiembre suele ser menor que en el resto del año por varias razones:

- Hay menos tráfico en la ciudad debido a las vacaciones de verano y a la menor actividad laboral y escolar.
- La radiación solar es más intensa y favorece la formación de ozono, que a su vez contribuye a reducir el dióxido de nitrógeno.
- Los vientos son más frecuentes y ayudan a dispersar los contaminantes.

```python
# Create a line graph for the average pollutants per hazard level (October to December)

start_date = pd.to_datetime('2022-10-01')
end_date = pd.to_datetime('2022-12-31')

date_range_data = df_date[(df_date['FECHA'] >= start_date) & (df_date['FECHA'] <= end_date)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("Dates from October to Decemberr")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

Los meses con menor contaminación en Madrid son los comprendidos entre octubre y diciembre por varias razones:

- Hay una mayor precipitación y humedad que favorecen la limpieza del aire y la eliminación de los contaminantes.
- Hay una menor radiación solar y una menor temperatura que reducen la formación de ozono troposférico.
- Hay una menor actividad económica y social que disminuye las emisiones de gases y partículas.

Sin embargo, esto no significa que la calidad del aire sea óptima en esos meses, ya que pueden producirse episodios de inversión térmica que atrapan los contaminantes cerca del suelo4.

Por eso, es importante seguir adoptando medidas para mejorar la calidad del aire durante todo el año.
