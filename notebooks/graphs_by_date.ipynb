# **Danger Graphs by Date**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_date = pd.read_csv('../data/processed/clean_M_date.csv')
```

```python
df_date.head()
```

```python
# Convert column 'FECHA' to 'datetime'
df_date['FECHA'] = pd.to_datetime(df_date['FECHA'])
```

```python
# Create a line chart with different colours for each hazard level (Full Year)

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_date, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("Full Year")
plt.ylabel("Pollutant Level")
plt.title("Evolution of Pollutants by Danger Level Over Time")

plt.xticks([])
plt.legend(title='Danger Level')
plt.show()
```

```python
# Create a line graph for the average pollutants per hazard level (First half of the year)

start_date_6 = pd.to_datetime('2022-01-01')
end_date_6 = pd.to_datetime('2022-05-30')

date_range_data_6 = df_date[(df_date['FECHA'] >= start_date_6) & (df_date['FECHA'] <= end_date_6)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data_6, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("First half of the year")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

La contaminacin en Madrid el 15 de marzo de 2022 fue bastante alta.

Algunas de las posibles causas de esta situacin son:

- El aumento del trfico rodado debido al fin de las restricciones por la pandemia de COVID-19.
- La falta de lluvias y viento que favorecen la dispersin de los contaminantes.

La contaminacin en los meses de entre junio y septiembre suele ser menor que en el resto del ao por varias razones:

- Hay menos trfico en la ciudad debido a las vacaciones de verano y a la menor actividad laboral y escolar.
- La radiacin solar es ms intensa y favorece la formacin de ozono, que a su vez contribuye a reducir el dixido de nitrgeno.
- Los vientos son ms frecuentes y ayudan a dispersar los contaminantes.

```python
# Create a line graph for the average pollutants per hazard level (October to December)

start_date = pd.to_datetime('2022-10-01')
end_date = pd.to_datetime('2022-12-31')

date_range_data = df_date[(df_date['FECHA'] >= start_date) & (df_date['FECHA'] <= end_date)]

sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=date_range_data, x='FECHA', y='MEDIA_DIARIA', hue='PELIGROSIDAD', palette='viridis')

plt.xlabel("Dates from October to Decemberr")
plt.ylabel("Average Pollutants")
plt.title("Evolution of the Pollutant Average by Danger Level")

plt.xticks(rotation=45)
plt.legend(title='Danger Level')
plt.show()
```

Los meses con menor contaminacin en Madrid son los comprendidos entre octubre y diciembre por varias razones:

- Hay una mayor precipitacin y humedad que favorecen la limpieza del aire y la eliminacin de los contaminantes.
- Hay una menor radiacin solar y una menor temperatura que reducen la formacin de ozono troposfrico.
- Hay una menor actividad econmica y social que disminuye las emisiones de gases y partculas.

Sin embargo, esto no significa que la calidad del aire sea ptima en esos meses, ya que pueden producirse episodios de inversin trmica que atrapan los contaminantes cerca del suelo4.

Por eso, es importante seguir adoptando medidas para mejorar la calidad del aire durante todo el ao.
